###  The "Create Feature" function will add extra columns relating to time

```python
def create_time_features(df):
    """
    Create time series features based on time series index.
    """
    df = df.copy()
    #df['hour'] = df.index.hour
    #df['dayofweek'] = df.index.dayofweek
    #df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    #df['dayofyear'] = df.index.dayofyear
    df['dayofmonth'] = df.index.day
    #df['weekofyear'] = df.index.isocalendar().week
    return df
```

### Determine the date to split the dataset into training and testing 

```python
# split train (75%) and test (25%). 
splitDate = '2022-09-01'  
```

# Build a model to predict Ontario's cases

```python
#---------------------Get X and y for Ontario------------------------
X = df.loc[df.date < splitDate]
y = df.loc[df.date >= splitDate]
```

```python
# let the date be the index so we can plot them
df = df.set_index("date") 
X = X.set_index("date") 
y = y.set_index("date")
```

```python
X.info()
```

```python
#---------------------Plot the X and y timeline of covid cases for Ontario------------------------
fig, ax = plt.subplots(figsize=(15, 5))
X['totalcases'].plot(ax=ax, label='Training Set (X)', title='Data Train (X) and Test (y) Split for Ontario')
y['totalcases'].plot(ax=ax, label='Test Set (y)')
ax.axvline(splitDate, color='black', ls='--')
ax.legend(['Training Set (X)', 'Test Set (y)'])
ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
plt.show()
```

```python
#---------------------Use the function create_time_features(dataframe) on df, X, and y------
df = create_time_features(df)
X = create_time_features(X)
y = create_time_features(y)
TARGET = "totalcases"  # our target variable
FEATURES = list(df.columns)
```

```python
df.info()
```

```python
#---------------------Train using cross validation--------------------------------
X_train = X[FEATURES]
y_train = X[TARGET]
X_test = y[FEATURES]
y_test = y[TARGET]
```

```python
#---------------------Train the model--------------------------------
reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',    
                       n_estimators=10000,
                       early_stopping_rounds=30,
                       objective='reg:squarederror',
                       max_depth=3,
                       gamma=0.4,
                       learning_rate=0.011)

reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        verbose=100)
```

```python
# save the predictions into a column, and merge it with df
y['prediction'] = reg.predict(X_test)  # add a column called 'predictions' to y
df = df.merge(y[['prediction']], how='left', left_index=True, right_index=True)
```

```python
#-------------------------Plot the prediction v.s actual----------------------------
ax = df[['totalcases']].plot(figsize=(15, 5))
df['prediction'].plot(ax=ax, style='.')
plt.legend(['Truth Data', 'Predictions'])
ax.set_title('Raw Data and Prediction')
plt.show()
```

```python
#--------------------Print out the RMSE score-------------------------------
score = RMSE(y['totalcases'], y['prediction'])
print(f'RMSE Score on Test set (predicting cases in Ontario): {score:0.2f}')
```
